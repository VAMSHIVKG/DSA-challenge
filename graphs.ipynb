```python
### krushkals using the disjiont operations

def find_parent(parent, i):
    # Function to find the parent of a node in a disjoint set
    if parent[i] == i:
        # print(i)
        return i
    
    parent[i] = find_parent(parent, parent[i])  # Path compression for optimization
    # print(i)
    return parent[i]

def union(parent, rank, x, y):
    # Function to perform union operation of two sets
    x_root = find_parent(parent, x)
    # print(x_root)
    y_root = find_parent(parent, y)
    # print(y_root)

    if x_root != y_root:  # If x and y don't belong to the same set
        if rank[x_root] < rank[y_root]:
            parent[x_root] = y_root
        elif rank[x_root] > rank[y_root]:
            parent[y_root] = x_root
        else:
            parent[y_root] = x_root
            rank[x_root] += 1

def kruskal(graph, V):
    # Function to find the minimum spanning tree using Kruskal's algorithm
    result = []
    graph = sorted(graph, key=lambda item: item[2])  # Sort edges by weight
    print("Sorted graph",graph)
    parent = [i for i in range(V)]  # Initialize parent array for disjoint sets
    print(parent)
    rank = [0] * V  # Initialize rank array for union-by-rank optimization
    print(rank)

    for u, v, weight in graph:  # Iterate through sorted edges
        print(u,v)
        x = find_parent(parent, u)
        print("x--->",x)
        y = find_parent(parent, v)
        print("y--->",y)
        if x != y:  # If adding this edge doesn't create a cycle
            result.append([u, v, weight])  # Add edge to the result
            union(parent, rank, x, y)  # Perform union operation

    minimum_cost = sum(edge[2] for edge in result)  # Calculate total cost of MST
    return result, minimum_cost  # Return MST edges and total cost

# Example usage:
graph = [
    [0, 1, 10],
    [0, 2, 6],
    [0, 3, 5],
    [1, 3, 15],
    [2, 3, 4]
]
V = 4

minimum_spanning_tree, minimum_cost = kruskal(graph, V)
print("Minimum Spanning Tree:")
for u, v, weight in minimum_spanning_tree:
    print(f"{u} -- {v} == {weight}")
print(f"Minimum Spanning Tree Cost: {minimum_cost}")




```
