```python


##left and right view

class Node:
    def __init__(self,d):
        self.left = None
        self.right = None
        self.d = d
        self.hd = 0

#constructing the binary tree
## LEFT VIEW tree statically
def Binary(nl):
    root = Node(nl[0])
    root.left = Node(nl[1])
    root.right = Node(nl[2])
    root.left.left = Node(nl[3])
    root.left.right = Node(nl[4])
    root.left.right.left = Node(nl[5])
    root.right.left = Node(nl[6])
    root.right.right = Node(nl[7])
    return root

def Left_Right_view(root,side):
    m={}
    q=[root]
    while q:
        node = q.pop(0)
        if node.hd not in m:
            m[node.hd] = []
        m[node.hd].append(node.d)
        if node.left:
            node.left.hd = node.hd + 1
            q.append(node.left)
        if node.right:
            node.right.hd = node.hd + 1
            q.append(node.right)

    print("LEFT AND RIGHT VIEW")
    if side=='L':
        print('\nLEFTVIEW')
        for hd in sorted(m.keys()):
            print(m[hd][0], end=' ')
    elif side=='R':
        print('\nRIGHTVIEW')
        for hd in sorted(m.keys()):
            print(m[hd][-1], end=' ')

    else:
        print("Nodes at each level")
        for hd in sorted(m.keys()):
            print(m[hd], end=' ')
        
            
            
        


## Driver code
nl=['A','B','C','D','E','F','G','H']
root=Binary(nl)
side=input()
Left_Right_view(root,side)
```

```python
##BOUNDARY TRAVERSAL
# in anti-clock direction
# starting from the root
# it includes the left boundary (excluding the leaf nodes)
# the leaf  nodes from the left to  right,
# and the right boundary (excluding the leaf nodes) in that order



##clock wise
class TreeNode:
    def __init__(self,d):
        self.left = None
        self.right = None
        self.d = d

def print_left_boundary(root):
    if root:
        if root.left:
            print_left_boundary(root.left)
            print(root.d,end='')
        else:
            print_right_boundary(root.right)
            print(root.d)

def print_leaves(root):
    if root:
        print_leaves(root.right)
        if  not root.left and not root.right:
            print(root.d)
        print_leaves(root.left)

def print_right_boundary(root):
    if root:
        if root.right:
            print(root.d)
            print_right_boundary(root.right)
        elif root.left:
            print(root.d)
            print_right_boundary(root.left)

def boundary_traversal(root):
    if not root:
        return
    print(root.d)
    print_right_boundary(root.right)
    print_leaves(root.right)
    print_leaves(root.left)
    print_left_boundary(root.left)

# Example usage:
# Construct the binary tree
root = TreeNode('A')
root.left = TreeNode('B')
root.left.left = TreeNode('C')
root.left.right = TreeNode('D')
root.left.right.left = TreeNode('E')
root.left.right.right = TreeNode('F')
root.right = TreeNode('G')
root.right.right = TreeNode('H')

# Perform boundary traversal
print("Boundary Traversal of the Binary Tree:")
boundary_traversal(root)
        
    
            
```

```python
##CLOCK WISE  BOUNDARY TRAVERSAL

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def print_leaves(node):
    if node:
        print_leaves(node.left)
        if not node.left and not node.right:
            print(node.val)
        print_leaves(node.right)

def print_left_boundary(node):
    if node:
        if node.left:
            print(node.val)
            print_left_boundary(node.left)
        elif node.right:
            print(node.val)
            print_left_boundary(node.right)

def print_right_boundary(node):
    if node:
        if node.right:
            print_right_boundary(node.right)
            print(node.val)
        elif node.left:
            print_right_boundary(node.left)
            print(node.val)

def boundary_traversal(root):
    if not root:
        return
    print(root.val)
    print_left_boundary(root.left)
    print_leaves(root.left)
    print_leaves(root.right)
    print_right_boundary(root.right)

# Example usage:
# Construct the binary tree
root = TreeNode('A')
root.left = TreeNode('B')
root.left.left = TreeNode('C')
root.left.right = TreeNode('D')
root.left.right.left = TreeNode('E')
root.left.right.right = TreeNode('F')
root.right = TreeNode('G')
root.right.right = TreeNode('H')

# Perform boundary traversal
print("Boundary Traversal of the Binary Tree:")
boundary_traversal(root)
```

```python

##level order traversal of a tree

#definind the node of the tree
class Node:
    def __init__(self,d):
        self.left = None
        self.right = None
        self.d = d
        # self.hd = 0

#constructing the binary tree
def Binary(nl):
    root = Node(nl[0])
    root.left = Node(nl[1])
    root.right = Node(nl[2])
    root.left.left = Node(nl[3])
    root.left.right = Node(nl[4])
    root.right.left = Node(nl[5])
    root.right.right = Node(nl[6])
    return root


def Level_order(root):
    m={}
    q=[(root,0)]
    while q:
        curr, lvl = q.pop(0)
        if lvl not in m:
            m[lvl] = []

        m[lvl].append(curr.d)

        if curr.left:
            q.append((curr.left, lvl + 1))

        if curr.right:
            q.append((curr.right, lvl + 1))

    print("LEVEL ORDER TRAVERSAL")
    for lvl in sorted(m.keys()):
        print(m[lvl], end=' ')    
    
    
    
    

## Driver code
nl=['A','B','C','D','E','F','G']
root=Binary(nl)
Level_order(root)
```

```python

##vertical order traversal of a tree

#definind the node of the tree
class Node:
    def __init__(self,d):
        self.left = None
        self.right = None
        self.d = d
        self.hd = 0

#constructing the binary tree
def Binary(nl):
    root = Node(nl[0])
    root.left = Node(nl[1])
    root.right = Node(nl[2])
    root.left.left = Node(nl[3])
    root.left.right = Node(nl[4])
    root.right.left = Node(nl[5])
    root.right.right = Node(nl[6])
    return root


def Vertical_order(root):
    m={}
    q=[(root,0)]
    while q:
        curr, hd = q.pop(0)
        if hd not in m:
            m[hd] = []

        m[hd].append(curr.d)

        if curr.left:
            q.append((curr.left, hd - 1))

        if curr.right:
            q.append((curr.right, hd + 1))

    print("VERTICAL ORDER TRAVERSAL")
    for hd in sorted(m):
    # for hd in sorted(m,reverse=True): for printing the nodes from bottom to top
        print(m[hd], end=' , ')    
    
    
    
    

## Driver code
nl=['A','B','C','D','E','F','G']
root=Binary(nl)
Vertical_order(root)
```
